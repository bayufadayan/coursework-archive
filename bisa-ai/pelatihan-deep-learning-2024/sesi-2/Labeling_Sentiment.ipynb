```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=",")

# Display the DataFrame
display(df)

# Check the length of the DataFrame
num_rows = len(df)
print(f"The number of rows in the DataFrame is: {num_rows}")
```

```python
# prompt: cleaning data from regular expression

import re

def clean_text(text):
  """Cleans the text by removing URLs, mentions, hashtags, and special characters."""
  text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
  text = re.sub(r'@\w+', '', text)  # Remove mentions
  text = re.sub(r'#\w+', '', text)  # Remove hashtags
  text = re.sub(r'[^\w\s]', '', text)  # Remove special characters
  text = re.sub(r'\s+', ' ', text).strip()  # Remove extra whitespace
  return text

# Apply the clean_text function to the 'full_text' column
data.apply(clean_text)

# Display the DataFrame with the cleaned text
data
```

```python
from textblob import TextBlob # Make sure you import the TextBlob library

data = df['full_text']

# prompt: cleaning data from regular expression

import re

def clean_text(text):
  """Cleans the text by removing URLs, mentions, hashtags, and special characters."""
  text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
  text = re.sub(r'@\w+', '', text)  # Remove mentions
  text = re.sub(r'#\w+', '', text)  # Remove hashtags
  text = re.sub(r'[^\w\s]', '', text)  # Remove special characters
  text = re.sub(r'\s+', ' ', text).strip()  # Remove extra whitespace
  return text

# Apply the clean_text function to the 'full_text' column
data = data.apply(clean_text) # Reassign data to the cleaned text

# Calculate the polarity for each text entry
data = data.apply(lambda x: TextBlob(x).sentiment.polarity) # Calculate polarity using TextBlob

data # Display the results
```

```python
# prompt: join polarity and full_text

# Assuming 'data' contains the polarity scores and 'df' is your original DataFrame
df['polarity'] = data

# You can now access the 'full_text' and 'polarity' columns together
df[['full_text', 'polarity']]
```

```python
df['polarity'].unique()
```

```python
# prompt: category 0 to minus with label  negatif

def categorize_polarity(polarity):
  if polarity < 0:
    return "negatif"
  else:
    return "positif"

df['sentiment'] = df['polarity'].apply(categorize_polarity)
df[['full_text', 'polarity', 'sentiment']]
```

```python
df['sentiment'].unique()
```

```python
# prompt: create word cloud for positif sentiment

!pip install wordcloud

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Filter the DataFrame for positive sentiments
positive_df = df[df['sentiment'] == 'positif']

# Combine all the positive tweets into a single string
text = " ".join(positive_df['full_text'].astype(str))

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the generated image:
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
